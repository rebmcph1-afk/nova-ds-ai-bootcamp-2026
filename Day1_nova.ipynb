## What Counts as Data?
Data is any collection of facts, measurements, or observations.

Examples:
- Numbers (temperature, age, income)
- Text (tweets, reviews)
- Images (X-rays, satellite photos)
- Audio (speech, music)
- Sensor readings (heart rate, GPS)

Data becomes useful only when we transform it into **information**.

## The Data Science / Machine Learning Lifecycle

1. Collect data  
2. Clean and prepare data  
3. Explore and visualize  
4. Build models  
5. Evaluate models  
6. Deploy models  
7. Monitor and update  

Humans are essential at every step.

## Categories of Machine Learning

### Supervised Learning
Model learns from labeled examples.
Examples: predicting house prices, classifying emails as spam.

### Unsupervised Learning
Model finds patterns without labels.
Examples: clustering customers, anomaly detection.

### Generative AI
Model creates new content.
Examples: ChatGPT, image generation, synthetic data.

# **Functions**
Define Function with an Argument

```python
def even_odd(num):
    if(num%2==0): #argument of the function if the number is a factor(%) of 2 and does not equal (==) 0
        return 'Even number'
    else:
        return 'Odd number'
```
