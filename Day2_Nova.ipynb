## 📁 Saving Files to GitHub

You can save notebooks and datasets to GitHub so they are easy to access from Colab.

---

### 🔼 Save a Notebook Directly from Colab

1. Open your notebook  
2. Go to **File → Save a copy in GitHub**  
3. Select your repository and folder  
4. Add a commit message  
5. Click **OK**

Your `.ipynb` file is now stored in GitHub.

---

### 🔼 Upload Files Manually (CSV, Excel, etc.)

1. Go to your GitHub repository  
2. Click **Add file → Upload files**  
3. Select your file (CSV, XLSX, JSON, etc.)  
4. Click **Commit changes**

Your dataset is now saved in GitHub and can be loaded from Colab

## 📊 Exploring Distributions in the California Housing Dataset

Below are two example plots showing how to visualize distributions in the California housing dataset. These help us understand how values are spread across the dataset.

### 🔹 Distribution of Median House Value

```python

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df['median_house_value'].hist(bins=40, color='skyblue', edgecolor='black')
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value")
plt.ylabel("Count")
plt.show()
````
